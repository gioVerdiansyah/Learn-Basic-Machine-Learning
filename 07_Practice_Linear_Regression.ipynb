**Latihan SKLearn Linear Regression**

**Tujuan** <br>
Memprediksi harga rumah berdasarkan jumlah kamar


**Tahap Latihan** <br>
Tahapan latihan yang akan dilakukan:
1. Import library yang dibutuhkan.
2. Buat datasets *dummy* deengan Numpy Array.
3. Buat plot dari model.

Import library yang di perlukan, dan membuat data dummy mengguankan NumPy array

```python
import numpy as np

# Data kamar
bedrooms = np.array([1,1,2,2,3,4,4,5,5,5])

# Data harga rumah, asumsikan dalam dollar
house_price = np.array([15000, 18000, 27000, 34000, 50000, 68000, 65000, 81000, 85000, 90000])
```

Selanjutnya, kita bisa mencoba menampilkan data tersebut dalam bentuk *scatter plot*. Jumlah kamar pada sumbu X adalah variabel independen dan harga rumah pada sumbu Y  adalah variabel dependen.

```python
# Menampilkan scatter dari plot dataset
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(bedrooms, house_price)
```

Lalu kita bisa melatih modedl kita dengan memanggil fungsi LinearRegression.fit(). Fungsi ini melatih model regresi linier dati *library SKLearn*.

```python
from sklearn.linear_model import LinearRegression

# latih model dengan fungsi LinearRegression.fit()
bedrooms = bedrooms.reshape(-1, 1)
linreg = LinearRegression()
linreg.fit(bedrooms, house_price)
```

Terakhir kita bisa melihat bagaimana model kita menyesuaikan dengan data yang kita miliki dengan membuat plot dari model kita.

```python
# Menampilkan plot hubungan antara jumlah kamar dengan harga rumah
plt.scatter(bedrooms, house_price)
plt.scatter(bedrooms, linreg.predict(bedrooms))
```

Model regresi linier adalah salah satu model machine learning yang paling sederhana. Model ini memiliki kompleksitas rendah dan bekerja sangat baik pada dataset yang memiliki hubungan linier.
