```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../resources/datasets/world_population.csv")
df
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
plt.figure(figsize=(22, 8))
sns.heatmap(df.isnull().T, cbar=False, cmap=['#201E43', "#FFB200"])
plt.title("Heatmap missing Value (null)")
plt.show()
```

```python
df.isnull().sum().plot(kind='bar', figsize=(22, 8))
plt.title("Null Total static", fontdict=dict({"fontsize": 20, "fontweight": "bold"}))
plt.xticks(rotation=45, ha='right', fontsize=12, fontweight="bold")

for i, value in enumerate(df.isnull().sum()):
    plt.text(i, value, str(value), ha='center', va='center', fontdict=dict({"fontsize": 15, "fontweight": "bold", "color": "#ff000d"}))

plt.show()
```

```python
import missingno as msno

msno.matrix(df, color=(0.1, 0.6, 0.8))
plt.show()
```
